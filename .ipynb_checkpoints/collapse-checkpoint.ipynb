```python
def gamma(R,m,U): 
  return np.sqrt(1+U**2-2*m / R)
 
def rho(R,m,U) : 
  return (gamma(R,m,U) + U) / (4*np.pi*R**2*dfdA(R))

def P(R,m,U,vs2) : 
  return vs2*rho(R,m,U)

def dfdA(f):
  return (np.concatenate((
    [ -25/12*f[0] + 4*f[1] - 3*f[2] + 4/3*f[3] - 1/4*f[4] ],
    [ -25/12*f[1] + 4*f[2] - 3*f[3] + 4/3*f[4] - 1/4*f[5] ],
     1/12*f[0:-4] - 2/3*f[1:-3] + 2/3*f[3:-1] - 1/12*f[4:],
    [ 25/12*f[-2] - 4*f[-3] + 3*f[-4] - 4/3*f[-5] + 1/4*f[-6] ],
    [ 25/12*f[-1] - 4*f[-2] + 3*f[-3] - 4/3*f[-4] + 1/4*f[-5] ]
  ))/deltaL)*np.exp(-L)

def psi(R,m,U,g,r,p) : 
  return np.log(g[-1]+U[-1]) + integ.cumtrapz(dfdA(U)/g + m/(4*np.pi*r*(R**4)*g) + p/(r*g*R), A, initial=0)

def k_coeffs(R,m,U,vs2) :
  g = gamma(R,m,U)
  r = rho(R,m,U)
  p = vs2 * r
  ep = np.exp(psi(R,m,U,g,r,p))
  
  #print(ep)
  kR = ep*U
  km = -1/ep*4*np.pi*R**2*p*U
  #kU = -ep/(1-vs2)*( 4*np.pi*g*R**2*r/(r + p)*dfdA(p) + (m + 4*np.pi*R**3*p)/R**2 ) \
  #  - ep*vs2/(1-vs2)*(4*np.pi*r*R**2*dfdA(U) + 2*U*g/R)
  kU = 1
  return kR, km, kU

```

We can set fairly generic initial conditions for matter with a constant equation of state (constant $v_s$) by choosing $\Gamma$, choosing $R = m/c(A)$ for a function $c(A=0) = 0$ while ensuring that $\rho$ is positive, and  solving for $U$, $m$, $\rho$ and $\psi$:

$$
{\rm choose\,\,}\Gamma(A)\,,\\
R(A) = m(A) / c(A) \,,\,{\rm and}\\
0 \le v_s^2 = {\rm const} \le 1/3 \,.\\
{\rm Then,}\\
U(A) = -\sqrt{ \Gamma(A)^2 - 1 + 2 c(A) }\\
m(A) = \int_0^A dA' \left( \Gamma(A') - v_s^2 U(A') \right) \\
\rho(A) = \frac{\Gamma + U}{4\pi R^2 R'}\\
\psi(A) = \ln(\Gamma + U)|_{A=A_{max}} + \int_0^A dA' \left( \frac{U'(A')}{\Gamma(A')} + \frac{m(A')}{4\pi \rho(A') R(A')^4 \Gamma(A') } \right)
$$

These expressions may contain divergences, but seem to be relatively stable when evolved.
